Question 1: What is hypothesis testing in statistics?

Answer:  Hypothesis testing is a formal statistical procedure used to evaluate claims or assumptions about a population based on sample data. It helps determine whether observed differences or effects are likely real or simply due to random chance.

Null Hypothesis (H‚ÇÄ): The default assumption that there is no effect or no difference.
Example: "The average website visits are 50 per day."

Alternative Hypothesis (H‚ÇÅ): The competing claim suggesting there is an effect or difference.
Example: "The average website visits are not 50 per day."

Significance Level (Œ±): The threshold for deciding whether to reject H‚ÇÄ, often set at 0.05 (5%).

p-value: The probability of observing the data (or something more extreme) if H‚ÇÄ is true. A small p-value (< Œ±) suggests rejecting H‚ÇÄ.

Test Statistic: A value calculated from sample data that is compared against critical values to decide whether to reject H‚ÇÄ.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:  In hypothesis testing, we always work with two competing statements about a population. Understanding their roles is crucial for interpreting statistical results.

Null Hypothesis (H‚ÇÄ)
Definition: The default assumption that there is no effect, no difference, or no relationship.

Form: Usually written with equality signs ( =, ‚â•, ‚â§ ).

Purpose: Serves as the baseline claim to be tested directly with data.

Example:

H‚ÇÄ: The average salary of data scientists is ‚Çπ10 LPA.

Alternative Hypothesis (H‚ÇÅ or H‚Çê)
Definition: The rival claim suggesting there is an effect, difference, or relationship.

Form: Written with inequality signs ( ‚â†, <, > ).

Purpose: Represents what the researcher wants to prove or provide evidence for.

Example:

H‚ÇÅ: The average salary of data scientists is not ‚Çπ10 LPA.

H‚ÇÅ: Model A and Model B have different accuracies.

Question 4: What are Type I and Type II errors? Give examples of each.

Answer:  When we perform hypothesis testing, there are two kinds of mistakes we can make. These are called Type I and Type II errors.

ype I Error (False Positive)
Definition: Rejecting the null hypothesis (H‚ÇÄ) when it is actually true.

Interpretation: You conclude there is an effect or difference, but in reality, there isn‚Äôt.

Probability: Equal to the significance level (Œ±).

Example:

A medical test says a patient has a disease (rejecting H‚ÇÄ: ‚Äúno disease‚Äù), but the patient is actually healthy.

A company concludes a new marketing strategy increases sales, but in reality, sales haven‚Äôt changed.

Type II Error (False Negative)
Definition: Failing to reject the null hypothesis (H‚ÇÄ) when it is actually false.

Interpretation: You conclude there is no effect or difference, but in reality, there is.

Probability: Denoted by Œ≤.

Example:

A medical test says a patient is healthy (fail to reject H‚ÇÄ: ‚Äúno disease‚Äù), but the patient actually has the disease.

A company concludes a new machine is no better than the old one, but in reality, it is more efficient.

Type I error is controlled by the significance level (Œ±).

Type II error is related to the test‚Äôs power (1 ‚àí Œ≤), which measures the ability to detect real effects.


Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:  A Z-test and a T-test are both statistical methods used in hypothesis testing to compare sample means, but they differ in the assumptions they make about the population and the sample size. A Z-test is typically used when the population variance (or standard deviation) is known and the sample size is large (generally ùëõ > 30). It relies on the standard normal distribution (Z-distribution) to evaluate whether the difference between sample and population means‚Äîor between two sample means‚Äîis statistically significant. In contrast, a T-test is used when the population variance is unknown and must be estimated from the sample, which is especially important when dealing with small sample sizes ( ùëõ < 30 ). The T-test relies on the t-distribution, which adjusts for the added uncertainty of estimating variance from limited data.

In practice, if you have a large dataset and know the population standard deviation, a Z-test is appropriate. However, in most real-world scenarios, the population variance is not known, and sample sizes are often modest, making the T-test the more commonly applied method. The choice between the two ensures that the statistical test accounts for the level of uncertainty in the data, with the T-test being more conservative for small samples and unknown variances, while the Z-test is more straightforward when population parameters are already established.

```python
# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
                        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu0 = 50  # hypothesized population mean
alpha = 0.05  # significance level

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu0) / (sample_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Sample Std Dev:", sample_std)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision
if p_value < alpha:
    print("Reject H0: The sample mean is significantly different from", mu0)
else:
    print("Fail to reject H0: No significant difference from", mu0)
```

```python
# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # mean and standard deviation
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# Step 2: Calculate 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
alpha = 0.05

# Standard error
se = sample_std / np.sqrt(n)

# Confidence interval using t-distribution
ci_lower, ci_upper = stats.t.interval(1 - alpha, df=n-1, loc=sample_mean, scale=se)

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))

# Step 3: Plot the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation

    # Compute Z-scores
    z_scores = [(x - mean) / std for x in data]

    # Print summary
    print("Mean:", mean)
    print("Standard Deviation:", std)
    print("First 5 Z-scores:", z_scores[:5])

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.legend()
    plt.show()

    return z_scores

# Example usage
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

z_scores = calculate_zscores(data)
```
